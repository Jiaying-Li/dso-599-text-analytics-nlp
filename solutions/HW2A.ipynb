```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import sys
sys.getdefaultencoding()
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
```

```python
# open file (combine poor and good)
review = open("amazon_toy_reviews.txt", "r")
```

```python
lines_review = review.readlines()
lines_review
```

```python
# tokenization and lemmatization
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer

lemmatizer = WordNetLemmatizer()
def stem(lines_review):
    sentence1 =[]
    for sentence in lines_review:
        token_words=word_tokenize(sentence)
        token_words
        stem_sentence=[]
        for word in token_words:
            stem_sentence.append(lemmatizer.lemmatize(word))
            stem_sentence.append(" ")
        sentence1.append("".join(stem_sentence))
    return sentence1

lines_review = stem(lines_review)

print(lines_review[2])
print(type(lines_review))
```

```python
#remove stop words
vectorizer = CountVectorizer(stop_words="english")
X = vectorizer.fit(lines_review)
X = vectorizer.fit_transform(lines_review)
```

```python
vectorizer.get_feature_names()
len(vectorizer.get_feature_names())
```

```python
import pandas as pd
review_df = pd.DataFrame(X.toarray(), columns = vectorizer.get_feature_names())
review_df.head()
```

```python
#sample 5000 reviews from all reviews
sample_5000 = review_df.sample(5000)
```

```python
sample_5000.sum(axis = 0, skipna = True) 
```

```python
reviews = pd.DataFrame(sample_5000.sum(axis = 0, skipna = True), columns = ['count'])
```

```python
# sort the word count by descending order
reviews_sort = reviews.sort_values(by='count', ascending=False)
reviews_sort.head()
```

```python
# reset index
reviews_sort= reviews_sort.reset_index()
reviews_sort.head()
```

```python
# reset columns' names
reviews_sort.columns = ['word', 'count']
reviews_sort.head()
```

```python
# find the top 500 words by count
top500 = reviews_sort.loc[ 0:499 , : ]
```

# Cosine Similarity

```python
# further sample 200 reviews from top500_data (5000 reviews with top 500 words by counts)
sample = top500_data.sample(200)
sample.head()
```

```python
sample.shape
```

```python
import numpy as np; import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(sample)
```

```python
# cosine similarity between each row
cosine_simi = pd.DataFrame(cosine_similarity(sample))
```

```python
cosine_simi.shape
```

```python
cosine_simi.head()
```

```python
untack_simi = cosine_simi.unstack()
```

```python
sorted_untack_simi = untack_simi.sort_values(ascending = False, kind = "quicksort")
sorted_untack_simi.head()
```

```python
sorted_untack_simi_df = sorted_untack_simi.reset_index()
sorted_untack_simi_df.head()
```

```python
# final result
# We can further look at original data to find this two reviews
sorted_untack_simi_df.loc[sorted_untack_simi_df['level_0'] != sorted_untack_simi_df['level_1']].head(1)
```

```python
# further explore the result
# we have many reviews that contain the exactly same content
sorted_untack_simi_df.loc[sorted_untack_simi_df['level_0'] != sorted_untack_simi_df['level_1']].head(50)
```
